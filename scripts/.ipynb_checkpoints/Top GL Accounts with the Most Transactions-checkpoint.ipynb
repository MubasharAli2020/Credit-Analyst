```python
#  Loading Required Libraries and Data

import pandas as pd
from IPython.display import display

# file path
file_path = "../data/Common Data Warehouse-Orginal.xlsx"

xls = pd.ExcelFile(file_path)

financial_df = xls.parse("Financial")  # Contains financial transactions
gl_df = xls.parse("GL")  # Contains GL Account details

#  Merging Financial Data with GL Table

financial_gl_df = financial_df.merge(gl_df, left_on="GL Acct ID", right_on="GL Account ID", how="left")

#  Counting Transactions by GL Account Name 

gl_account_transactions = financial_gl_df.groupby("GL Name").agg(
    Transaction_Count=("GL Acct ID", "count"),
    Total_Amount=("Amount", "sum")  # Summing transaction amounts to see financial movements
)

# Sorting by highest transaction count
gl_account_transactions = gl_account_transactions.sort_values(by="Transaction_Count", ascending=False)

#  Results
print("ðŸ”¹ Top GL Accounts with the Most Transactions:")
display(gl_account_transactions)

#  Save Results to Excel 
output_file_path = "../data/Top_GL_Accounts_With_Most_Transactions.xlsx"
gl_account_transactions.to_excel(output_file_path, index=False)

print(f"Results saved successfully to: {output_file_path}")
```
