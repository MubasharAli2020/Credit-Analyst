```python
import pandas as pd
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from scripts.load_data import load_data
```

```python
#  file path
file_path = "../data/Common Data Warehouse-Orginal.xlsx"

final_df = load_data(file_path)

# Displaying first few rows
final_df.head()
```

```python
# Creating a new column to identify whether the transaction is Adjusted (ADJ) or Original
final_df["Adjustment_Flag"] = final_df["Source"].astype(str).apply(lambda x: "Adjusted" if "ADJ" in x else "Original")

# Displaying unique values to verify
final_df["Adjustment_Flag"].value_counts()
```

```python
# Grouping by Sector Name and Adjustment Flag
sector_adjustment_analysis = final_df.groupby(["Sector Name", "Adjustment_Flag"]).agg(
    Total_Amount=("Amount", "sum")
).reset_index()

# Pivoting for better visualization
sector_adjustment_pivot = sector_adjustment_analysis.pivot(
    index="Sector Name", columns="Adjustment_Flag", values="Total_Amount"
).reset_index().fillna(0)

sector_adjustment_pivot.columns = ["Sector Name", "Total Amount Before Adjustment", "Total Amount After Adjustment"]

# results
sector_adjustment_pivot
```

```python
# Save results to an Excel file
sector_adjustment_pivot.to_excel("data/Sector_Adjustment_Analysis.xlsx", index=False)
print("Sector Adjustment Analysis saved to data/Sector_Adjustment_Analysis.xlsx")
```
