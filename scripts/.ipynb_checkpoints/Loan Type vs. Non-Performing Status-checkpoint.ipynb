```python
# Loading Required Libraries

import pandas as pd
from IPython.display import display

# file path
file_path = "../data/Common Data Warehouse-Orginal.xlsx"

xls = pd.ExcelFile(file_path)

instrument_df = xls.parse("Instrument")  # Contains loan type and performance status

#  Counting Non-Performing Loans by Loan Type 

# Grouping by Agreement Purpose (Loan Type) and count Performing/Non-Performing status
loan_performance = instrument_df.groupby(["Agreement Purpose", "Performing/Non Performing"]).agg(
    Loan_Count=("Agreement ID", "count")
).reset_index()

# Pivot the Data

loan_performance_pivot = loan_performance.pivot(index="Agreement Purpose", columns="Performing/Non Performing", values="Loan_Count").fillna(0)

# Renaming the columns
loan_performance_pivot.columns = ["Performing", "Non-Performing"]

# Adding a column for Non-Performing
loan_performance_pivot["Non-Performing Rate (%)"] = (loan_performance_pivot["Non-Performing"] / (loan_performance_pivot["Performing"] + loan_performance_pivot["Non-Performing"])) * 100

# Sorting by highest non-performing rate
loan_performance_pivot = loan_performance_pivot.sort_values(by="Non-Performing Rate (%)", ascending=False)

# Results ---
print("ðŸ”¹ Loan Type vs. Non-Performing Status:")
display(loan_performance_pivot)

#  Save Results to Excel
output_file_path = "Loan_Type_vs_Non_Performing_Status.xlsx"
loan_performance_pivot.to_excel(output_file_path, index=True)

print(f"Results saved successfully to: {output_file_path}")
```
